# Introdução.

#### As [distribuições normais](https://www.wikiwand.com/en/Normal_distribution) são frequentemente usados nas ciências naturais e sociais para representar variáveis aleatórias de valor real cujas distribuições não são conhecidas. A distribuição normal é uma distribuição de probabilidade teórica contínua.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

## Função de densidade de probabilidade de distribuição normal (Normal Distribution [Probability Density Function](https://www.investopedia.com/terms/p/pdf.asp)).

$
\text{a Função densidade de probabilidade (Probability density function - PDF) da distribuição normal é:} \\
$
$$
f(x)=\frac{1}{\sigma \sqrt{2 \pi}}e^{-\frac{(x- \mu)^2}{2 \sigma^2}}\text{, em que $x\in \mathbb{R}$}
$$

A variável aleatória $ X $ descrita pelo PDF é normal que segue uma distribuição normal com média $ \mu $ e variância $ \sigma^{2} $.
$
\text{Normal distribution notation is} \\
$
$$
X \sim N(\mu,\sigma^2) \\
$$

A [área sob a curva]((https://mathworld.wolfram.com/RiemannSum.html)) é igual à $ 1 $.
$$
\int \limits_{-\infty}^{+\infty} f(x) dx = 1
$$

## O método [`norm.pdf`].

#### O método [`norm.pdf`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html) retorna um valor $ pdf $. A seguir vemos o $ pdf $ quando $x = 1$, $\mu = 0$, $\sigma = 1$.

```python
norm.pdf(x = 1.0, 
         loc = 0, 
         scale = 1)
```

```python
fig, ax = plt.subplots()
x = np.arange(-4, 4, 0.001)

ax.plot(x, 
        norm.pdf(x)
       )

ax.set_ylim(0, 0.45) # range

ax.axhline(y = 0.24, 
           xmax = 0.61, 
           color = 'r'
          ) # horizontal line

ax.axvline(x = 1, 
           ymax = 0.53, 
           color = 'r', 
           alpha = 0.5
          ) # vertical line

xplot = ax.plot([1], 
                [0.24], 
                marker = 'o', 
                markersize = 15, 
                color = "red"
               ) # coordinate point

ax.set_yticks([]) # remove y axis label

ax.set_xticks([]) # remove x axis label

ax.set_xlabel('x', 
              fontsize = 20
             ) # set x label

ax.set_ylabel('pdf(x)', 
              fontsize = 20, 
              rotation = 0 
             ) # set y label

ax.xaxis.set_label_coords(0.61, 
                          -0.02
                         ) # x label coordinate

ax.yaxis.set_label_coords(-0.1, 
                          0.5 
                         ) # y label coordinate

plt.show()
```


#### Uma vez que [`norm.pdf`] retorna um valor PDF, podemos usar esta função para plotar a função de distribuição normal. Representamos graficamente um PDF da distribuição normal usando as bibliotecas [`scipy`](https://www.scipy.org/), [`numpy`](https://numpy.org/) e [`matplotlib`](https://matplotlib.org/). 

#### Usamos o domínio $ -4 < x <4 $, intervalo $ 0 <f (x) < 0,45 $, padrão $ \mu = 0 $ e $ \sigma = 1 $. `plot(xvalues, yvalues)` para produz o gráfico. 

```python
fig, ax = plt.subplots()

x = np.arange(-4, 4, 0.001)

ax.set_title('N(0,$1^2$)')

ax.set_xlabel('x')

ax.set_ylabel('f(x)')

ax.plot(x, 
        norm.pdf(x)
       )

ax.set_ylim(0, 0.45)

plt.show()
```

#### Uma curva normal tem a forma de um sino suave, é simétrica em relação a $ x = \mu $ e tem um ponto máximo em $ x = \mu $.

## PDF de uma distribuição normal, para diferentes valores de desvio padrão.

#### O método `scipy.norm.pdf` tem o parâmetro `loc`, que especifica a média e o parâmetro `scale`. que especifica o desvio padrão. Vamos plotar funções de distribuição de probabilidade de uma distribuição normal em que a média tem diferentes desvios-padrão.

```python
fig, ax = plt.subplots()

x = np.linspace(-10, 10, 100)

stdvs = [1.0, 2.0, 3.0, 4.0]

for s in stdvs:
    ax.plot(x, 
            norm.pdf(x, scale = s), 
            label = 'stdv=%.1f' % s
           )
    
ax.set_xlabel('x')

ax.set_ylabel('pdf(x)')

ax.set_title('Normal Distribution')

ax.legend(loc = 'best', 
          frameon = True
         )

ax.set_ylim(0, 0.45)

ax.grid(True)
```

## PDF de uma distribuição normal, para diferentes valores de média.

#### Vamos plotar funções de distribuição de probabilidade de distribuição normal onde o desvio padrão tem valor $ 0 $ e médias diferentes.

```python
fig, ax = plt.subplots()

x = np.linspace(-10, 10, 100)

means = [0.0, 1.0, 2.0, 5.0]

for mean in means:
    ax.plot(x, 
            norm.pdf(x, loc = mean), 
            label = 'mean=%.1f' % mean
           )
    
ax.set_xlabel('x')

ax.set_ylabel('pdf(x)')

ax.set_title('Normal Distribution')

ax.legend(loc = 'best', 
          frameon = True
         )

ax.set_ylim(0,0.45)

ax.grid(True)
```

#### A média da distribuição determina a localização do centro do gráfico. Como você pode ver no gráfico acima, a forma do gráfico não muda com a mudança da média, mas o gráfico é transladado horizontalmente.

# Usando valores de distribuição normal aleatória.


#### O método [`norm.rvs`](https://docs.scipy.org/doc/scipy-0.16.1/reference/generated/scipy.stats.norm.html) gera números de distribuição normal aleatória de acordo com o desvio padrão `scale`, a média `loc`. Criamos um histograma para os números gerados e adicionamos o $ pdf $.

```python
fig, ax = plt.subplots()

xs = norm.rvs(scale = 2,
              size = 1000
             )

x = np.linspace(-10, 10, 100)

p = norm.pdf(x, 
             scale = 2
            )

v = np.var(xs)

m = np.mean(xs)

#ax = fig.add_subplot(111)

ax.hist(xs, 
        bins = 10, 
        alpha = 0.5, 
        density = True
       )

ax.plot(x, 
        p, 
        'r-', 
        lw = 2
       )

ax.set_xlabel('x')

ax.set_ylabel('pdf(x)')

ax.set_title(f'mean={m:.2f}, var={v:.2f}')

ax.grid(True)
```

# Função de distribuição normal cumulativa ([Cumulative normal distribution function](https://www.allaboutcircuits.com/technical-articles/cumulative-distribution-function-probabilities-normally-distributed-data/#:~:text=The%20Cumulative%20Normal%20Distribution%20Function&text=This%20means%20that%20the%20Gaussian,taken%20along%20the%20Gaussian%20curve.))

#### A função de distribuição cumulativa de uma variável aleatória $ X $, avaliada em $ x $, é a probabilidade de $ X $ assumir um valor menor ou igual a $ x $. Como a distribuição normal é uma distribuição contínua, a área sombreada da curva representa a probabilidade de $ X $ ser menor ou igual a $ x $.

$$
P(X \leq x)=F(x)=\int \limits _{-\infty} ^{x}f(t)dt \text{, em que }x\in \mathbb{R}
$$

#### Usando o método [`.fill_between()`](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.fill_between.html) preenchemos a área entre duas curvas `y1` e` y2`.

```python
fig, ax = plt.subplots()

# for distribution curve

x = np.arange(-4, 4, 0.001)

ax.plot(x, 
        norm.pdf(x)
       )

ax.set_title("Distribuição normal cumulativa")

ax.set_xlabel('x')

ax.set_ylabel('pdf(x)')

ax.grid(True)

# for fill_between

px = np.arange(-4, 1, 0.01)

ax.set_ylim(0, 0.5)

ax.fill_between(px, 
                norm.pdf(px), 
                alpha = 0.5, 
                color = 'g'
               )

# for text

ax.text(-1, 0.1, "cdf(x)", fontsize = 20)

plt.show()
```

# Cálculo da probabilidade da distribuição normal.

#### Dada a média de $ 3 $ e o desvio padrão de $ 2 $, podemos encontrar a probabilidade de $ P (X  < 2) $.

$$
X \sim N(3, 2^{2}) 
$$

#### No método [`norm.cdf()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), o parâmetro localização (`loc`) especifica a média the mean e o parâmetro escala (`scale`) especifica o desvio padrão.

```python
from scipy.stats import norm

lessthan2 = norm.cdf(x = 2, 
                     loc = 3, 
                     scale = 2
                    )
print(lessthan2)
```

#### Plotamos o gráfico da distribuição.

```python
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-4, 10, 0.001)

ax.plot(x, norm.pdf(x, 
                    loc = 3, 
                    scale = 2
                   )
       )

ax.set_title("N(3,$2^2$)")

ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# for fill_between
px = np.arange(-4, 2, 0.01)
ax.set_ylim(0, 0.25)
ax.fill_between(px, 
                norm.pdf(px,loc = 3, scale = 2), 
                alpha = 0.5, 
                color = 'g'
               )

# for text
ax.text(-0.5, 0.02, round(lessthan2, 2), fontsize = 20)
plt.show()
```

## Intervalo entre variáveis.

To find the probability of an interval between certain variables, you need to subtract `cdf` from another `cdf`. Let's find $P(0.5<X<2)$ with a mean of 1 and a standard deviation of 2.


#### Para encontrar a probabilidade de um intervalo entre certas variáveis, você precisa subtrair `cdf` de outro` cdf`. Vamos encontrar $ P (0,5 < X <2) $ com uma média de $ 1 $ e um desvio padrão de $ 2 $.

$$ 
X \sim N(1, 2^{2}) 
$$

```python
norm(1, 2).cdf(2) - norm(1,2).cdf(0.5)
```

A seguir a curva para a probabilidade de encontrarmos um valor $(0.5 < X < 2)$, <br/>  

$$ 
X \sim N(1,2^2), P(0.5 < X < 2)
$$

```python
fig, ax = plt.subplots()

# for distribution curve
x = np.arange(-6, 8, 0.001)
ax.plot(x, norm.pdf(x,
                    loc = 1, 
                    scale = 2)
       )

ax.set_title("N(1,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)


px = np.arange(0.5, 2, 0.01)
ax.set_ylim(0, 0.25)
ax.fill_between(px,norm.pdf(px,loc = 1, scale = 2), 
                alpha = 0.5, color = 'g'
               )

pro = norm(1, 2).cdf(2) - norm(1,2).cdf(0.5)
ax.text(0.2, 0.02, round(pro, 2), fontsize = 20)
plt.show()
```

#### Para encontrar a probabilidade de $ P(X > 4) $ podemos usar a [função de sobrevivência](https://towardsdatascience.com/survival-functions-101-bd57fb8be888) (também  definida como $ 1 - cdf$), calculada pelo método [`norm.sf()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), que retorna a probabilidade $ P(X > 4) $ em que: 

$$
$\mu = 4, \text{e}~ \sigma = 2
$$

```python
gr4sf = norm.sf(x = 4, 
                loc = 3, 
                scale = 2
               )
gr4sf
```

#### Plotamos a seguir esse gráfico.

```python
fig, ax = plt.subplots()

x = np.arange(-4, 10, 0.001)

ax.plot(x, norm.pdf(x, loc = 3, scale = 2))
ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

px=np.arange(4,10,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px, 
                norm.pdf(px,loc = 3, scale = 2), 
                alpha = 0.5, color = 'g'
               )

ax.text(4.5, 0.02, "sf(x) %.2f" %(gr4sf), fontsize = 20)
plt.show()
```

O gráfico acima é idêntico àquele do tipo $ 1 - P(X < 4) $.

```python
gr4 = norm.cdf(x = 4, 
               loc = 3, 
               scale = 2
              )
gr14 = 1 - gr4

fig, ax = plt.subplots()
x = np.arange(-4, 10, 0.001)

ax.plot(x, norm.pdf(x, 
                    loc = 3, 
                    scale = 2)
       )

ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

px = np.arange(4, 10, 0.01)
ax.set_ylim(0, 0.25)
ax.fill_between(px,norm.pdf(px, 
                            loc = 3, 
                            scale = 2), 
                alpha = 0.5, 
                color = 'g'
               )
px1=np.arange(-4, 4, 0.01)
ax.fill_between(px1, 
                norm.pdf(px1, loc = 3, scale = 2), 
                alpha = 0.5, 
                color= 'r'
               )

ax.text(4.5, 0.02, round(gr14, 2), fontsize = 20)
ax.text(1,0.02, round(gr4, 2), fontsize = 20)
plt.show()
```

# Encontrando valores de quantis.

#### A propabilidade $P(X \leq k) = 0.95$ de um valor de $ X \leq k$ ocorrer caracteriza o [quartil](https://www.statisticshowto.com/what-are-quartiles/#:~:text=Find%20Quartiles%3A%20Examples&text=Step%201%3A%20Put%20the%20numbers,3%20numbers%20in%20each%20quartile.) de $ 95 \% $.

## Função de ponto percentual

#### o método [`norm.ppf()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html) é o inverso de `cdf` e é chamado de [função de ponto percentual](https://machinelearningmastery.com/continuous-probability-distributions-for-machine-learning/#:~:text=The%20inverse%20of%20the%20CDF,or%20equal%20to%20a%20probability.). Dada a média de $ 1 $ e o desvio padrão de $ 3 $, podemos encontrar o quantil `a` em $ P (X < a) = 0,506 $ usando `ppf`.


$$
X \sim N(1, 3^{2})
$$

```python
norm.ppf(q = 0.506, 
         loc = 1, 
         scale = 3
        )
```

```python
fig, ax = plt.subplots()

x = np.arange(-10, 10, 0.001)

ax.plot(x, 
        norm.pdf(x, loc = 1, scale = 3)
       )

ax.set_title("N(1,$3^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

xpoint = norm.ppf(q = 0.506, 
                  loc = 1, 
                  scale = 3
                 )

px = np.arange(-10, xpoint, 0.01)
ax.set_ylim(0,0.15)
ax.fill_between(px, 
                norm.pdf(px, 
                         loc = 1, 
                         scale = 3), 
                alpha = 0.5, 
                color = 'g'
               )

ax.text(.8, 0.02, "x= %.2f" %xpoint, fontsize = 20)
ax.text(-5, 0.05, "P(X)=0.506", fontsize = 20)
plt.show()
```

## Função de sobrevivência inversa

#### Com a mesma média e desvio padrão, podemos encontrar o quantil `b` em $ P (X > b) = 0.198 $ usando a [função sobrevivência inversa](https://www.itl.nist.gov/div898/handbook/eda/section3/eda362.htm#:~:text=The%20survival%20function%20is%20the,a%20value%20greater%20than%20x.&text=Just%20as%20the%20percent%20point,of%20the%20percent%20point%20function.), calculada com o método [`norm.isf()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Isso é o mesmo que usar `ppf` com $ q = (1 - 0.198) $.

$$
X \sim N(1,3^{2})
$$

```python
norm.isf(q = 0.198, loc = 1, scale = 3)
```

```python
norm.ppf(q = (1 - 0.198), loc = 1, scale = 3)
```

## Intervalo em torno da média.

#### O método [`norm.interval`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html) retorna pontos de extremidade do intervalo que contém a porcentagem alfa da distribuição. Por exemplo, com uma média de $ 0 $ e um desvio padrão de $ 1 $ para encontrar $ 95 \% $ da probabilidade, `norm.interval` retorna $ x $ valores em torno da média, neste caso, $ \mu = 0 $.

$$ 
X \sim N(0, 1^{2})
$$

```python
a, b = norm.interval(alpha = 0.95, loc = 0, scale = 1)
print(a, b)
```

```python
fig, ax = plt.subplots()

x = np.arange(-4, 4, 0.001)

ax.plot(x, norm.pdf(x))

ax.set_title("Interval")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

px = np.arange(a, b, 0.01)
ax.set_ylim(0, 0.5)
ax.fill_between(px, 
                norm.pdf(px), 
                alpha = 0.5, 
                color = 'g'
               )

ax.text(-0.5, 0.1, "0.95", fontsize = 20)
plt.show()
```

# Distribuição Normal Multivariada.

#### A distribuição normal multivariada é freqüentemente usada para descrever qualquer conjunto de variáveis aleatórias correlacionadas de valor real. 

#### Usamos a biblioteca [`multivariate_normal`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multivariate_normal.html) que usa uma matriz diagonal em que todos os elementos fora da diagonal são zero

```python
from scipy.stats import multivariate_normal

x, y = np.meshgrid(np.linspace(-10, 10, 100), 
                  np.linspace(-10,10,100)
                 )

pos = np.dstack((x, y))

mean = np.array([1, 2])
cov  = np.array([[3,0], [0,15]])
rv = multivariate_normal(mean, cov)

z = rv.pdf(pos)

fig = plt.figure()
ax = fig.add_subplot(111, aspect = 'equal')

ax.contourf(x, y, z)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('pdf')
```

# Distribuição normal padrão.

#### Quando $ \mu = 0 $ e $ \text{variância} = 1 $, é chamada de [distribuição normal padrão](https://statisticsbyjim.com/basics/normal-distribution/). A função de probabilidade acima é simplificada para:

$$
\Phi(x)=\frac{1}{\sqrt{2\pi}}e^{\frac{-x^2}{2}}\\
Z \sim N(0,1)
$$

#### Todas as curvas normais podem ser relacionadas à distribuição normal padrão.

## Variável normal padronizada.

#### Para padronizar uma variável aleatória $X \sim N(\mu, \sigma^2)$ como uma variável normal $Z \sim N(0,1)$, usamos a transformação:

$$
Z = \frac{x-\mu}{\sigma}
$$

#### O valor padronizado $ Z $ informa quantos desvios padrão abaixo ou acima da média é o valor original.

## Encontrando o valor normal padrão.

#### Por exemplo, para encontrar o valor padronizado para $ x = 1 $ quando a média é de $ \mu = 2 $ e o desvio padrão de $ \sigma = 3 $.

$$
\text{For }X \sim N(2, 3^2)\\
Z=\frac{1-2}{3}
$$

```python
z = (1 - 2) / 3
z
```

#### Podemos usar o método [`norm.cdf`](https://docs.scipy.org/doc/scipy/reference/tutorial/stats.html) para encontrar uma função de distribuição cumulativa e o método [`norm.ppf()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html) para encontrar a função de ponto percentual, [inversa](https://towardsdatascience.com/an-introduction-to-statistical-analysis-and-modelling-with-python-ef816b67f8ff) à [`cdf`].

```python
norm.cdf(1, loc = 2, scale = 3)
```

```python
norm.ppf(q = norm.cdf(1, loc = 2, scale = 3))
```
