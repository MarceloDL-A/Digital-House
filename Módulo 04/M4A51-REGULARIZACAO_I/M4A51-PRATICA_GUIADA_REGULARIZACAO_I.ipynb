# PRÁTICA GUIADA: Regularização em Machine Learning

#### Um dos principais aspectos do treinamento do seu modelo de aprendizado de máquina é evitar o [`overfitting`](https://elitedatascience.com/overfitting-in-machine-learning), pois neste caso o modelo terá uma baixa precisão. Isso acontece porque o seu modelo dificilmente irá conseguir capturar o ruído em seu conjunto de dados de treinamento. Por [`ruído`](https://towardsdatascience.com/how-to-use-noise-to-your-advantage-5301071d9dc3), queremos dizer os pontos de dados que realmente não representam as propriedades reais de seus dados, mas de chance aleatória. Aprender esses pontos de dados torna o seu modelo mais flexível, sob o risco de [`overfitting`](https://machinelearningmastery.com/introduction-to-regularization-to-reduce-overfitting-and-improve-generalization-error/).

#### O conceito de [`balanceamento`](https://towardsdatascience.com/regularization-the-path-to-bias-variance-trade-off-b7a7088b4577) de [`viés`](https://community.datarobot.com/t5/resources/regularization-and-the-bias-variance-tradeoff-part-1/ta-p/7029) e [`variância`](https://medium.com/analytics-vidhya/bias-variance-tradeoff-9ca44703bf7c) é útil para entender o fenômeno do [`overfitting`](https://datascience.foundation/sciencewhitepaper/underfitting-and-overfitting-in-machine-learning) e variação de balanceamento, para controlar erros do Machine Leanrning.

## Formas de Regularização

#### Vamos abordar aqui duas formas de se fazer a regularização:

    I. Aumentar o conjunto de dados (número de observações);
    II. Penalização dos otimizadores.

## Aumento do conjunto de dados

#### O [`overfitting`](https://towardsdatascience.com/overfitting-vs-underfitting-a-complete-example-d05dd7e19765) pode ser controlado aumentando o tamanho do conjunto de dados de treinamento. Aumentar o tamanho do conjunto de dados para evitar ajustes excessivos se refere ao aumento do número de observações (ou linhas) e não do número de atributos (ou colunas). A adição de colunas pode levar ao aumento da complexidade do problema e, portanto, pode resultar em um pior desempenho. A seguir veremos um exemplo de como aumentar um conjunto de dados para Regularizar um modelo

## Penalização de otimizadores

#### [`Regularização`](https://towardsdatascience.com/what-are-overfitting-and-underfitting-in-machine-learning-a96b30864690) é uma técnica que restringe/regulariza/reduz as estimativas dos coeficientes, diminuiindo seus valores, que ficam portanto mais próximos de zero. Em outras palavras, esta técnica desencoraja a aprendizagem de um modelo mais complexo ou flexível, de modo a evitar o risco de [`overfitting`](https://datascience.foundation/sciencewhitepaper/underfitting-and-overfitting-in-machine-learning).

#### Uma relação simples de [`regressão linear`](https://towardsdatascience.com/regularized-linear-regression-models-57bbdce90a8c) se parece com a fórmula abaixo, onde $y$ representa a relação aprendida e $\theta$ representa as estimativas de coeficiente para diferentes variáveis ou preditores $x$.

<br>
<br>
$$
y = \theta_0 + \theta_1x_1 + \theta_2x_2 + \theta_3x_3 + ... + \theta_nx_n
$$
<br>
<br>

#### O procedimento de ajuste envolve uma [`função de perda`](https://towardsdatascience.com/regularization-an-important-concept-in-machine-learning-5891628907ea), conhecida como [`soma residual de quadrados`]() ou RSS, sendo que os coeficientes são escolhidos, de forma que minimizem essa função de perda.

#### A [`regularização`](https://www.futurelearn.com/info/courses/advanced-machine-learning/0/steps/49535) ajustará os [`coeficientes`](https://towardsdatascience.com/regularization-in-machine-learning-76441ddcf99a#:~:text=This%20is%20a%20form%20of,linear%20regression%20looks%20like%20this.) com base nos seus dados de treinamento. Se houver ruído nos dados de treinamento, os coeficientes estimados não serão generalizados nos dados futuros. É aqui que a [`regularização`](https://analyticsindiamag.com/regularization-in-machine-learning-a-detailed-guide/) entra e encolhe, ou regulariza, essas estimativas aprendidas para zero.

## Tipos de Regularização por Penalização dos Otimizadores

## 1. Ridge Regression

<br>
<br>
$$
Ridge : \lambda \sum_{i=1}^{n} \theta_i^2
$$
<br>
<br>

#### A fórmula acima mostra o coeficiente de `Ridge` que modifica a regressão pela adição da quantidade de contração. Agora, os coeficientes são estimados minimizando essa função, e o $\lambda$ é o parâmetro de sintonia que decide quanto queremos penalizar a flexibilidade do nosso modelo. O aumento na flexibilidade de um modelo é representado pelo aumento de seus coeficientes e, se quisermos minimizar a função acima, esses coeficientes precisam ser pequenos. É assim que a técnica de regressão de [`Ridge`](https://towardsdatascience.com/ridge-regression-for-better-usage-2f19b3a202db) impede que os coeficientes subam demais. Além disso, observe que encolhemos a associação estimada de cada variável com a resposta, exceto o interceptador $\theta_0$. 

#### Quando $\lambda = 0$, o termo de penalidade não tem efeito, e as estimativas produzidas por [`regressão de rebordo`](http://rstudio-pubs-static.s3.amazonaws.com/385563_aff45a3836e149669652a12dc102a64b.html) serão iguais a mínimos quadrados. No entanto, como $\lambda \rightarrow \infty$, o impacto da penalidade de contração aumenta, e as estimativas coeficientes da regressão da crista se aproximam de zero. Como pode ser visto, selecionar um bom valor de $\lambda$ é crítico. A [`validação cruzada`](https://machinelearningmastery.com/ridge-regression-with-python/) é útil para esse propósito. As estimativas dos coeficientes produzidas por este método são também conhecidas como [`norma L2`](https://towardsdatascience.com/intuitions-on-l1-and-l2-regularisation-235f2db4c261).

#### Os coeficientes que são produzidos pelo método dos mínimos quadrados padrão são de escala equivariante, ou seja, se multiplicarmos cada entrada por $c$, então os coeficientes correspondentes são escalonados por um fator de $1 / c$. Portanto, independentemente de como o preditor é escalado, a multiplicação de preditor e coeficiente permanece a mesma. No entanto, esse não é o caso da regressão de `Ridge` e, portanto, precisamos padronizar os preditores ou trazer os preditores para a mesma escala antes de executar a regressão de `Ridge`.

## 2. Lasso Regression

<br>
<br>
$$
Lasso : \lambda \sum_{i=1}^{n} |\theta_i| 
$$
<br>
<br>

#### `Lasso` é outra variação de regularização, em que a função acima é minimizada. É claro que essa variação difere da regressão de `Ridge` apenas em penalizar os altos coeficientes. Ele usa $| \theta_i |$ (módulo) em vez de $\theta_i^{2}$ (quadrado), como sua penalidade. Nas estatísticas, isso é conhecido como a [`norma L1`](https://medium.com/@minions.k/lasso-regression-in-detail-l1-regularization-593044a85248).

#### Vamos dar uma olhada nos métodos acima com uma perspectiva diferente: a regressão de `Ridge` pode ser pensada como resolvendo uma equação, onde a soma dos quadrados dos coeficientes é menor ou igual a $s$. E o `Lasso` pode ser pensado como uma equação onde a soma do módulo de coeficientes é menor ou igual a $s$. Aqui, $s$ é uma constante que existe para cada valor do fator de encolhimento $\lambda$. Essas equações também são chamadas de funções de restrição.

#### Considere os seus dois parâmetros em um determinado problema. Então, de acordo com a formulação acima, a regressão de `Ridge` é expressa por 

<br>
<br>
$$
\theta_1^2 +  \theta_2^2 ≤ s.
$$
<br>
<br>

#### Isto implica que os coeficientes de regressão de `Ridge` têm a menor RSS (função de perda) para todos os pontos que se encontram dentro da circunferência dada por $\theta_1^2 +  \theta_2^2 ≤ s$.

#### Da mesma forma, para `Lasso`, a equação se torna 

<br>
<br>
$$ 
|\theta_1| +  |\theta_2| ≤ s.
$$
<br>
<br>

#### Isto implica que os coeficientes de `Lasso` possuem o menor RSS (função de perda) para todos os pontos que estão dentro do diamante dado por $ |\theta_1| +  |\theta_2| ≤ s $.

<br>
<img src="regularizacao_grafico.png" style="height:350px">
<br>

#### A imagem acima mostra as funções de restrição (áreas verdes), para Lasso (esquerda) e regressão de Ridge (direita), juntamente com contornos para RSS (elipse vermelha). Pontos na elipse compartilham o valor do RSS. Para um valor muito grande de $s$, as regiões verdes conterão o centro da elipse, fazendo estimativas de coeficiente de ambas as técnicas de regressão, iguais às estimativas de mínimos quadrados. 

#### Mas, este não é o caso na imagem acima. Nesse caso, as estimativas do coeficiente de regressão de `Lasso` e `Ridge` são fornecidas pelo primeiro ponto em que uma elipse entra em contato com a região de restrição. Como a regressão de `Ridge` possui uma restrição circular sem pontos agudos, essa interseção geralmente não ocorre em um eixo e, portanto, as estimativas do coeficiente de regressão de `Ridge` serão exclusivamente diferentes de zero. No entanto, a restrição de `Lasso` tem cantos em cada um dos eixos e, portanto, a elipse geralmente cruza a região de restrição em um eixo. Quando isso ocorre, um dos coeficientes será igual a zero. Em dimensões mais altas, onde os parâmetros são muito maiores que $2$, muitas das estimativas de coeficiente podem ser iguais a zero simultaneamente.

#### Isso esclarece a desvantagem óbvia da regressão de `Ridge`, que é a interpretabilidade do modelo. Reduzirá os coeficientes dos preditores menos importantes, muito próximos de zero. Mas isso nunca os tornará exatamente zero. Em outras palavras, o modelo final incluirá todos os preditores. No entanto, no caso do `Lasso`, a penalidade de `L1` tem o efeito de forçar algumas das estimativas de coeficiente a serem exatamente iguais a zero quando o parâmetro de ajuste $\lambda$ é suficientemente grande. Portanto, o método do `Lasso` também realiza [`seleção de variáveis`](https://medium.com/@23.sargam/lasso-regression-for-feature-selection-8ac2287e25fa) e é dito que produz modelos escassos.

## 3. ElasticNet Regression (L1 + L2)

#### A regressão [`ElasticNet`](https://machinelearningmastery.com/elastic-net-regression-in-python/) combina o poder da regressão `Ridge` e `Lasso` em um algoritmo. O que isso significa é que, com a `ElasticNet`, o algoritmo pode remover variáveis fracas como no `Lasso` ou reduzi-las a quase zero, como no `Ridge`. Todos esses algoritmos são exemplos de Regularização.

## Trade-off entre viés e variância

#### O modelo de regressão linear simples, no qual pretende-se predizer $n$ observações da variável de resposta, $Y$, com uma combinação linear de $m$ variáveis preditoras, $X$ e um termo de erro normalmente distribuído com variação $\simga^{2}$, é definido por uma equação linear. Como não conhecemos os parâmetros verdadeiros, temos que estimar esses parâmetros a partir da amostra. Na abordagem de mínimos quadrados ordinários (OLS), estimamos-os como $\beta$ estimados de tal maneira que a soma dos quadrados dos resíduos é a menor possível.

## O que a Regularização faz exatamente?

#### Um modelo padrão de mínimos quadrados tende a apresentar alguma variação, ou seja, esse modelo não generaliza bem para um conjunto de dados diferente dos dados de treinamento. A regularização reduz significativamente a variância do modelo, sem aumento substancial de seu viés. Portanto, o parâmetro de ajuste $\lambda$, usado nas técnicas de regularização descritas acima, controla o impacto na polarização e variação. 

#### À medida que o valor de $\lambda$ aumenta, reduz o valor dos coeficientes e, assim, reduz a variação. Até certo ponto, esse aumento em $\lambda$ é benéfico, pois reduz apenas a variação (evitando, portanto, o ajuste excessivo), sem perder propriedades importantes nos dados. Mas, após um certo valor, o modelo começa a perder propriedades importantes, dando origem a um viés no modelo e, portanto, a um ajuste insuficiente. Portanto, o valor de $\lambda$ deve ser cuidadosamente selecionado.

#### A [regularização](https://towardsdatascience.com/regularization-in-machine-learning-76441ddcf99a) uma técnica útil que pode ajudar a melhorar a precisão dos seus modelos de regressão. Uma biblioteca popular para implementar esses algoritmos é o [`Scikit-Learn`](https://medium.com/coinmonks/regularization-of-linear-models-with-sklearn-f88633a93a2).

## Regressão Ridge Teórica

#### Apresentaremos agora uma classe que realiza a Regularização de `Ridge` e faremos a aplicação deta regularização em dados sintéticos.

### Classe de Regularização Ridge

#### Nesta classe de modelo de regressão Ridge, podemos identificar o elemento `I` da Regularização, que é baseado no hiper-parâmetro`l` sendo adicionado na modelagem para regularizar o cálculo dos pesos ou estimadores `w_hat`, que são utilizados na predição.

```python
class ridge_regression(object):

    def __init__(self, l = 1e-4):
        self.l = l

    def fit(self, X_train, y_train):
        # adiciona coluna de 1 nos dados
        X = np.insert(X_train, 0, 1, 1)

        # estima os w_hat
        I = np.identity(X.shape[1]) * self.l # multiplicando o coeficiente de regularização l pela Identidade
        w_hat = np.dot( np.dot( np.linalg.inv(np.dot(X.T, X) + I), X.T), y_train) # termo de regularização: I
        
        # calcula os estimadores
        self.w_hat = w_hat
        self.coef = self.w_hat[1:]
        self.intercept = self.w_hat[0]

    def predict(self, X_test):
        X = np.insert(X_test, 0, 1, 1) # adiciona coluna de 1 nos dados
        y_pred = np.dot(X, self.w_hat) # X * w_hat = y_hat
        return y_pred
```

## Aplicação

#### Regularização de uma função de ajuste a dados sitéticos gerados a partir de uma função quadrática, onde o hiper-parâmetro de regularização é aumentado, generalizando a função de otimização de uma regressão com `overfit` para uma regressão genérica. o gráfico das fuuções de otimização com o aumento do hiper-parâmetro `l` demonstra como a otimização genérica fica livre da variância dos dados de treino.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
plt.figure(figsize = (20,6))

# definindo uma função quadrática para geração dos dados sintéticos
def f(x):
    return x**2 - 10*x + 5 + np.random.normal(0,10,size=x.shape)

# Gera os dados sintéticos com a função quadrática
x_train = np.random.uniform(0, 15, 15)
y = f(x_train)
X = x_train[:, np.newaxis]

# parâmetros do gráfico
colors = ['blue', 'green', 'orange']
lw = 2

# ajusta polinômios de grau 10 com diferentes coeficinetes de regularização
for count, l in enumerate([1e-8, 1e-3, 1e2]):
    
    # crias as variáveis polinomiais e as passa pelo modelo de regressão de ridge
    model = make_pipeline(PolynomialFeatures(10), 
                          ridge_regression(l = l)
                         )
    model.fit(x_train.reshape(-1, 1), y)
    
    # coloca as previsões no gráfico
    x_plot = np.linspace(0, 15, 50)
    y_plot = model.predict(x_plot.reshape(-1, 1))
    plt.plot(x_plot.reshape(-1, 1), 
             y_plot, 
             color = colors[count], 
             linewidth = lw,
             label = "$\lambda =$ %.1e" % l
            )

plt.scatter(x_train, 
            y, 
            color = 'navy', 
            s = 30, 
            marker = 'o'
           )
plt.ylim([-60, 90])
plt.legend(loc = 'best')
plt.show()
```

## Regularização 

### Sample dos dados

#### Inicilamente vamos diminuir a quantidade de dados por cinco, calcular uma regressão linear e medir os resultados de treino e teste (RMSE de treino e teste). Então calculamos a diferença para verificar a perda pelo sobreajuste dos dados. 

#### Depois disso, vamos trabalhar com o DataSet com o tamanho original, o que para nós representa uma regularização, e repetir o processo para comparar a diferença entre as diferenças de metrica entre treino e teste dos dados de Sample e dos dados completos ou Regularizados.

#### Esperamos que a diferença entre as métricas RMSE de treino e teste sejam menores para os dados regularizados, o que significa que temos um modelo mais genérico, ou seja, menos viciado nos dados de treino, ou com menos variância, e capaz de entender melhor os padrões dos dados de teste ou dados desconhecidos.

## Diminuindo o tamanho dos dados em n vezes

#### Retiramos uma amostra do `dataset`.

```python
n = 3
df_sample = df.sample(n = int(df.shape[0] / n), 
                      random_state = 0
                     )
df_sample.shape
```

### Split dos dados SAMPLE em treino e teste

```python
from sklearn.model_selection import train_test_split
# split dos dados
train, test = train_test_split(df_sample, test_size = 0.2, 
                               random_state = 0
                              )
```

### Ajuste do modelo SAMPLE por uma regressão linear

#### Criamos uma função com as variáveis e rodando uma regressão linear.

```python
function = 'PRICE ~ ZN + CHAS + RM + RAD + \
                    CRIM + AGE + PTRATIO + B + LSTAT + \
                    INDUS + NOX + DIS + TAX'

# instanciando, ajustando e imprimindo os resultados do modelo OLS
model = smf.ols(function, train).fit()
print(model.summary())
```

### Medindo os resultados SAMPLE com a diferença entre treino e teste

```python
# importando as bibliotecas
from sklearn.metrics import mean_squared_error
from math import sqrt

# fazendo predições de treino
y_hat_train = model.predict(train)
# calculando a raiz quadrada do erro médio quadrático
rmse_train = sqrt(mean_squared_error(train.PRICE, 
                                     y_hat_train
                                    )
                 )
print('RMSE de Treino:', rmse_train)

# fazendo predições de teste
y_hat_test = model.predict(test)
# calculando a raiz quadrada do erro médio quadrático
rmse_test = sqrt(mean_squared_error(test.PRICE, 
                                    y_hat_test
                                   )
                )
print('RMSE de Teste:', rmse_test)

print('Diferença:', rmse_test - rmse_train)
```

### Split dos dados REGULARIZADOS em treino e teste

```python
# split dos dados
train, test = train_test_split(df, test_size = 0.2, 
                               random_state = 99
                              )
```

### Ajuste do modelo REGULARIZADO por uma regressão linear

#### Criamos uma função com as variáveis e rodamos uma regressão linear, instanciando, ajustando e imprimindo os resultados do modelo `OLS`.

```python
function = 'PRICE ~ ZN + CHAS + RM + RAD + \
                    CRIM + AGE + PTRATIO + B + LSTAT + \
                    INDUS + NOX + DIS + TAX'


model = smf.ols(function, train).fit()
print(model.summary())
```

### Medindo os resultados REGULARIZADOS com a diferença entre treino e teste

```python
# fazendo predições de treino
y_hat_train = model.predict(train)

# calculando a raiz quadrada do erro médio quadrático
rmse_train = sqrt(mean_squared_error(train.PRICE, 
                                     y_hat_train
                                    )
                 )
print('RMSE de Treino:', rmse_train)

# fazendo predições de teste
y_hat_test = model.predict(test)
# calculando a raiz quadrada do erro médio quadrático
rmse_test = sqrt(mean_squared_error(test.PRICE, 
                                    y_hat_test
                                   )
                )
print('RMSE de Teste:', rmse_test)

print('Diferença:', rmse_test-rmse_train)
```

## Conclusão 

#### Como esperado, a diferença entre as métricas de treino e teste para o modelo REGULARIZADO foi menor, o que prova que o modelo REGULARIZADO tem maior capacidade de generalização.

### Criação de variáveis DUMMIES

#### Escolhendo uma lista de variáveis categóricas para a criação de `DUMMIES`.

```python
# criando uma lista com as variáveis categóricas candidatas a DUMMIES
categoricals = ['condition', 
                'num_floors', 
                'is_waterfront', 
                'c_num_bath', 
                'c_num_partial_bath', 
                'zip'
               ]
# Criando uma função que cria variáveis dummies das colunas escolhidas
def one_hot(df, cols):
    for each in cols:
        dummies = pd.get_dummies(df[each], 
                                 prefix = each, 
                                 drop_first = True
                                )
        df = pd.concat([df, dummies], 
                       axis = 1
                      )
    return df

# aplicando a função no DataFrame e nas colunas selecionadas
df = one_hot(df,categoricals)

# Eliminando as colunas originais
df = df.drop(categoricals, axis = 1)

# visualizando os dados
df.head()
```

#### Agroa vamos rodas os modelos de `Machine Learning` utilizando os pacotes de regressão de `Lasso` e `Ridge` para entender como estes algoritmos lidam com a regularização de parâmetros.

```python
# importando as bibliotecas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
```

### Separando os dados em treino e teste

```python
# função para separar os dados em treino e teste
def dfSplit(df, ratio, y = 'log_price'):
    train, test = train_test_split(df, 
                                   test_size = ratio, 
                                   random_state = 0
                                  )
    y_train = train[y]
    y_test = test[y]
    #x_train = train.ix[:, train.columns != y]
    #x_test = test.ix[:, test.columns != y]
    x_train = train.iloc[:, train.columns != y]
    x_test = test.iloc[:, test.columns != y]        
    return x_train, y_train, x_test, y_test
```

### Calculando variáveis polinomiais 

#### A Função polinomial tenta expandir o polinomio das colunas explicativas ou `X`. Isso serve para que a função possa entender comportamos não lineares. Se tivermos interações e polinomios sem necessidade, podemos causar `overfiting`, ou seja, superajustar nosso modelo.

```python
# função que calcula variáveis polinomiais
def tPoly(df, degree = 1):
    polynomial = PolynomialFeatures(degree = degree)
    return polynomial.fit_transform(df)
```

# Funções para rodar os modelos de Machine Learning

#### A função que roda os modelos de `machine learning`, aplica o polinomio, separa os dados em treino e teste, exibe as métricas e plota os resultados. 

#### Como o `Cross Validation Score` está sempre tentando minimizar o erro (mse), o resultado médio minimizado sempre acaba negativo, por isso, no momento de extrair a raiz (rmse), é necessário o [sinal negativo](https://github.com/scikit-learn/scikit-learn/issues/2439).

#### Definimos uma função que roda os modelos de `Machine Learning` e aplica as outras funções.

```python
 scorer = make_scorer(mean_squared_error, greater_is_better = False)
def testRegs(df, clf, degree = 1, ratio = 0.2, y = 'log_price', metrics = []):

    # aplicando a função para separação dos dados
    x_train, y_train, x_test, y_test = dfSplit(df, 
                                               ratio, 
                                               y = 'log_price'
                                              )
    # aplicando a função de cálculo de polinomiais nos dados de treino e teste
    poly_x_train = tPoly(x_train, 
                         degree
                        )
    poly_x_test = tPoly(x_test, 
                        degree
                       )
    # ajustando o modelo selecionado
    clf.fit(poly_x_train,y_train)
    
    # fazendo predições
    y_hat = clf.predict(poly_x_test)
    
    # calculando a raiz quadrada do erro médio quadrático RMSE (negativo pois o cross_val_score minimiza o valor resultante)
    rmse = np.sqrt(-cross_val_score(clf, 
                                    poly_x_train, 
                                    y_train, 
                                    scoring = scorer, 
                                    cv = 10
                                   )
                  )
    # imprimindo o RMSE
    print("RMSE:", rmse.mean())
    
    # fazendo predições com dados de treino e teste
    y_train_pred = clf.predict(poly_x_train)
    y_test_pred = clf.predict(poly_x_test)

    # imprimindo os resultados
    print('R2: %.2f, Score: %.2f, Parameters: %i' % (r2_score(y_test, y_hat), 
                                                     clf.score(poly_x_train,y_train), 
                                                     clf.coef_.shape[0]))
    # imprimindo os resultados
    #print('R2: %.2f, Score: %.2f, Parameters: %i' % (r2_score(y_test, y_hat), 
    #                                                 clf.score(poly_x_train,y_train), 
    #                                                 clf.coef_.shape[0]))
    
    # plotando uma dispersão entre os dados preditos de treino e teste pelos erros
    # para dados de treino em azul e dados de teste em verde
    plt.scatter(y_train_pred, 
                y_train_pred - y_train, 
                c = "blue", 
                marker = "s", 
                label = "Training data"
               )
    plt.scatter(y_test_pred, 
                y_test_pred - y_test, 
                c = "lightgreen", 
                marker = "s", 
                label = "Validation data"
               )
    plt.xlabel("Predicted values")
    plt.ylabel("Residuals")
    plt.hlines(y = 0, 
               xmin = 10.5, 
               xmax = 13.5, 
               color = "red"
              )
    plt.show()

    # plotando uma dispersão entre os dados preditos de treino e teste pelos dados de treino e teste
    # para dados de treino em azul e dados de teste em verde
    plt.scatter(y_train_pred, 
                y_train, 
                c = "blue", 
                marker = "s", 
                label = "Training data"
               )
    plt.scatter(y_test_pred, 
                y_test, 
                c = "lightgreen", 
                marker = "s", 
                label = "Validation data"
               )
    plt.xlabel("Predicted values")
    plt.ylabel("Real values")
    plt.legend(loc = "upper left")
    plt.plot([10.5, 13.5], 
             [10.5, 13.5], 
             c = "red"
            )
    plt.show()
```

### Definindo novas funções para as regressões

#### Definimos uma função para imprimir as métricas do modelo.

```python
def print_metrics(y_train, yhat_train,y_test, yhat_test):
    # imprimindo dados de treino
    print('\n-----Dados de Treino-----')
    print('RMSE - treino', np.sqrt(mean_squared_error(y_train, 
                                                      yhat_train
                                                     )
                                  )
         )
    print('R2 - treino', r2_score(y_train, 
                                  yhat_train
                                 )
         )
    # imprimindo dados de teste
    print('\n-----Dados de Teste-----')
    print('RMSE - test', np.sqrt(mean_squared_error(y_test, yhat_test)))
    print('R2 - test', r2_score(y_test, yhat_test))
    print('\n------------------------')
```

#### Definimos também uma função para rodar os modelos com funções polinomiais.

```python
def run_model_with_poly(dataframe, poly_n, modelo, target = 'log_price'):

    # instanciando a função polinomial    
    poly = PolynomialFeatures(degree = poly_n)

    # definindo oss dados e entrada
    X = df.drop([target], 
                axis = 1
               )

    # definindo a vairável target
    newY = dataframe[target]
    
    # criando um novo DataFrame com as variáveis polinomiais e imprimindo o novo shape
    df_temp = pd.DataFrame(poly.fit_transform(X))
    print('\ndf_temp shape', df_temp.shape)

    # separando os dados em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(df_temp, 
                                                        newY, 
                                                        test_size = 0.2, 
                                                        random_state = 0
                                                       )
    # imprimindo os shapes dos dados de treino e teste
    print('\ntrain shape', X_train.shape)
    print('teste shape', X_test.shape)

    # ajustando o modelo
    modelo.fit(X_train, y_train)
    
    # fazendo predições para treino e teste
    yhat = modelo.predict(X_train)
    yhat_test = modelo.predict(X_test[list(X_train)])
    
    # imprimindo as métricas
    print_metrics(y_train, 
                  yhat, 
                  y_test, 
                  yhat_test
                 )
    return modelo
```
