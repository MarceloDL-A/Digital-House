#### Exercício 2 - Faça uma seleção inicial do número de colunas que você imagina que possuam uma maior relevância para o `dataset` e realize uma contagem inicial do número de nulos e sua porcentagem no `dataset`, incluindo os tipos e a contagem de cada tipo de dado. Calcule também os valores de tend6encia central de cada coluna do `dataset`.

```python
def EDA_stats(df):
    df= df.rename(columns = str.lower)

    eda_df = {}
    eda_df['Amount_NaN'] = df.isnull().sum()
    eda_df['%_NaN'] = df.isnull().mean().round(2)
    eda_df['DType'] = df.dtypes
    eda_df['Amount_Data'] = df.count()
    
    colunas = df.columns.str.lower().tolist()
    
    eda_df['Mean'] = np.round(df.mean(), 
                              2
                             )
    
    eda_df['Median'] = np.round(df.median()
                                , 2
                               )
    
    eda_df['Max'] = df.max()
    eda_df['Min'] = df.min()
    eda_df['STD'] = np.round(df.std(), 2)
    
    eda = pd.DataFrame(eda_df)
    
    eda['Amount_Unique'] = list(map(lambda x: len(df[x].unique().tolist()), 
                                    colunas
                                   )
                               )

    return eda
```

```python
EDA_stats(IowaHousePricing)
```
